#### Using Free API 

```python
import requests
```

```python
def get_quotes():
    url = "https://api.freeapi.app/api/v1/public/quotes/quote/random"

    responce = requests.get(url)
    if responce.status_code == 200:
        data = responce.json()
        main = data["data"]["content"]

        return main
```

```python
# The command to get quotes 
data = get_quotes()
print(data)
```

### GET Todos

```python
def get_todos():
    url = "https://api.freeapi.app/api/v1/todos"
    data_raw = requests.get(url)
    if data_raw.status_code == 200:
        dataa = data_raw.json()
        message = dataa["message"]
        return message


todo = get_todos()
print(todo)
```
